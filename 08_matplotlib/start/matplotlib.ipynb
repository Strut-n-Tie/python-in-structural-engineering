## Line customisation

Optional parameters in `plot()` function.

| Parameter| Purpose | Command | 
|-------------------|---------|---------|
| **`color`** | Custom colours |`colour = "green"`, `color="k"`, `color='#2E86AB'` (hex), `color="C0"` |
| **`linewidth`** | Controls line thickness | `linewidth=2.5` | 
| **`alpha`** | Controls transparency | `alpha=0.5` | 
| **`marker`** | Point markers | `marker='o'`, `marker='s'` | 
| **`markersize`** | Marker size | `markersize='3'` | 
| **`linestyle`** | Line pattern | `linestyle='-.'`, `linestyle=':'`, `linestyle='--'` | 

List of available colours: https://matplotlib.org/stable/gallery/color/named_colors.html.
List of avialable markers: https://matplotlib.org/stable/api/markers_api.html.

`linewidth`, `marker` and/or `linestyle` can be replaced by an **`fmt`** function. Examples: `'b-'` (blue solid), `'r--'` (red dashed).
