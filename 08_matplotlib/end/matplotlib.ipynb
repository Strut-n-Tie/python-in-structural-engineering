# Matplotlib

```python
from matplotlib import pyplot as plt
```

## Basic plotting

```python
# Creating a canvas with pre-arranged subplots - better customisation

x = [0, 0.001, 0.15]
y = [0, 355, 510]

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(x, y)
ax.set_title('Steel stress-strain curve') 
ax.set_ylabel('Stress [MPa]') 
ax.set_xlabel('Strain [%]')
ax.set_xlim([0, None])
ax.set_ylim([0, None])
ax.grid(True, alpha=0.3)
plt.show()
```

```python
# Creating a blank canvas - simple plots (not used in this course)

x = [0, 0.001, 0.15]
y = [0, 355, 510]

plt.figure(figsize=(10, 4))
plt.plot(x, y)
plt.title('Steel stress-strain curve') 
plt.ylabel('Stress [MPa]') 
plt.xlabel('Strain [%]')
plt.xlim([0, None])
plt.ylim([0, None])
plt.grid(True, alpha=0.3)
plt.show()
```

## Legend

Location ``loc` parameters:
* `best` 
* `upper right`
* `upper left`
* `lower left`
* `lower right`
* `right`
* `center left`
* `center right`
* `lower center`
* `upper center`
* `center`

```python
x = [0, 0.001, 0.15]
y1 = [0, 355, 510]
y2 = [0, 235, 410]

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(x, y1, label = "S355 ($f_y = 355$ MPa)")
ax.plot(x, y2, label = "S3235 ($f_y = 235$ MPa)")
ax.set_title('Steel stress-strain curve') 
ax.set_ylabel('Stress [MPa]') 
ax.set_xlabel('Strain [%]')
ax.set_xlim([0, None])
ax.set_ylim([0, None])

# Default
ax.legend()
# Add location
ax.legend(loc='lower right')
# Add title
ax.legend(title='Steel Grades', title_fontsize=12)
# Split into different columns
ax.legend(ncol=2, columnspacing=1.5)
# Outside the plot - right-hand side
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
# Outside the plot - bottom
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2))

# Example customised legend with six labels
ax.plot(x, y2, label = "S3235 ($f_y = 235$ MPa)")
ax.plot(x, y2, label = "S3235 ($f_y = 235$ MPa)")
ax.plot(x, y2, label = "S3235 ($f_y = 235$ MPa)")
ax.plot(x, y2, label = "S3235 ($f_y = 235$ MPa)")
ax.legend(title='Steel Grades', title_fontsize=12,ncol=3, columnspacing=1.5,loc='upper center', bbox_to_anchor=(0.5, -0.2))


ax.grid(True, alpha=0.3)
plt.show()
```

## Line customisation

Optional parameters in `plot()` function.

| Parameter| Purpose | Command | 
|-------------------|---------|---------|
| **`color`** | Custom colours |`colour = "green"`, `color="k"`, `color='#2E86AB'` (hex), `color="C0"` |
| **`linewidth`** | Controls line thickness | `linewidth=2.5` | 
| **`alpha`** | Controls transparency | `alpha=0.5` | 
| **`marker`** | Point markers | `marker='o'`, `marker='s'` | 
| **`markersize`** | Marker size | `markersize='3'` | 
| **`linestyle`** | Line pattern | `linestyle='-.'`, `linestyle=':'`, `linestyle='--'` | 

List of available colours: https://matplotlib.org/stable/gallery/color/named_colors.html.
List of avialable markers: https://matplotlib.org/stable/api/markers_api.html.

`linewidth`, `marker` and/or `linestyle` can be replaced by an **`fmt`** function. Examples: `'b-'` (blue solid), `'r--'` (red dashed).

```python
# Creating a canvas with pre-arranged subplots - better customisation

x = [0, 0.001, 0.15]
y1 = [0, 355, 510]
y2 = [0, 235, 410]
y3 = [0, 510, 740]

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(x, y1, linestyle = "-", color = "mediumturquoise", linewidth = 3, label = "S355")
ax.plot(x, y2, linestyle = "--", color = "#AB2E8C", marker = "o", markersize = 5, linewidth = 1, label = "S235")
ax.plot(x, y3, "o:b", label = "S510") # markers and the line
ax.plot(x, y3, "ok", label = "S510") # markers only
ax.set_title('Steel stress-strain curve') 
ax.set_ylabel('Stress [MPa]') 
ax.set_xlabel('Strain [%]')
ax.set_xlim([0, None])
ax.set_ylim([0, None])
ax.grid(True, alpha=0.3)
plt.show()
```

## Annotations

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(x, y1)
y = [0, 410, 510]
ax.text(x[1]*1.1,y[1]*1.1, f"$f_y = {y[1]}$")
ax.text(x[2]*0.95,y[2]*0.9, f"$f_u = {y[2]}$")
ax.set_title('Steel stress-strain curve') 
ax.set_ylabel('Stress [MPa]') 
ax.set_xlabel('Strain [%]')
ax.set_xlim([0, None])
ax.set_ylim([0, None])
ax.grid(True, alpha=0.3)
plt.show()
```

## Multiple plots

`fig, axes = plt.subplots(no_rows, no_cols, figsize=(x, y))`

```python
fig, ax = plt.subplots(2,1,figsize=(7,7))

ax[0].grid(which='major', linestyle=':', linewidth='0.5', color='black')
ax[0].plot(x, y1, linestyle = "-", color = "mediumturquoise", linewidth = 3, label = "S355")
ax[0].set_title('S355') 
ax[0].set_ylabel('Stress [MPa]') 
ax[0].set_xlabel('Strain [%]')

ax[1].grid(which='major', linestyle=':', linewidth='0.5', color='black')
ax[1].plot(x, y2, linestyle = "--", color = "#AB2E8C", marker = "o", markersize = 5, linewidth = 1, label = "S235")
ax[1].set_title('S235') 
ax[1].set_ylabel('Stress [MPa]') 
ax[1].set_xlabel('Strain [%]')

# Fit the plots together
plt.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(3,3,figsize=(7,7))

ax[1][0].grid(which='major', linestyle=':', linewidth='0.5', color='black')
ax[1][0].plot(x, y2, linestyle = "--", color = "#AB2E8C", marker = "o", markersize = 5, linewidth = 1, label = "S235")
ax[1][0].set_title('S235') 
ax[1][0].set_ylabel('Stress [MPa]') 
ax[1][0].set_xlabel('Strain [%]')

# Fit the plots together
plt.tight_layout()

plt.show()
```
