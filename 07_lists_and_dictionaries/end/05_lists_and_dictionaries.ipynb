# Lists and Dictionaries

```python
materials = ["steel", "concrete", "timber"]
print(materials[2])
materials[2] = "aluminum"
print(materials[2])
```

```python
def propped_cantilever_reactions(w, L):
    R1 = 3*w*L/8
    R2 = 5*w*L/8
    return [R1, R2]

w = 30 # kN/m
L = 8 # m
reactions = propped_cantilever_reactions(w, L)
print(f"R1 = {reactions[0]} kN, R2 = {reactions[1]} kN")
```

```python
beam_length = 5
spacing = 0.5
number_of_nodes = int(beam_length / spacing + 1)
print(f"number of nodes = {number_of_nodes}")

nodes = []

for i in range(number_of_nodes):
    print(i)
    nodes.append(i * spacing)

print(nodes)
```

```python
beam_length = 5
spacing = 0.5
number_of_nodes = int(beam_length / spacing + 1)
print(f"number of nodes = {number_of_nodes}")

nodes = [0.1]

for i in range(1,number_of_nodes,2):
    print(i)
    nodes.append(i * spacing)

print(nodes)
```

```python
# List comprehension
nodes = [i for i in range(number_of_nodes)]
print(nodes)
```

```python
# List comprehension with an if-statement
# [expression for variable in iterable if condition]
nodes = [i for i in range(number_of_nodes) if i % 2 == 0]
print(nodes)
```

### List methods

`list.method(*inputs)`

| Method | Description |
|--------|-------------|
| `list.append(elem)` | Add a single element to the end of `list`. |
| `list.insert(index, elem)` | Insert an element at the given index. |
| `list.remove(elem)` | Remove the first occurrence of the element. |
| `list.pop(index)` | Remove and return the element at the given index. |
| `list.extend(list2)` | Add all elements from `list2` to the end of `list`. |
| `list.index(elem)` | Return the index of the first occurrence of the element. |
| `list.sort()` | Sort `list`. |
| `list.reverse()` | Reverse `list`. |

```python
# insert
numbers = [1, 3, 4]
numbers.insert(1, 2)

print(numbers)
```

```python
# remove
numbers = [1, 2, 3, 2]
numbers.remove(2)

print(numbers)
```

```python
# index
numbers = [1, 2, 3, 2]
idx = numbers.index(2)

print(idx)
```

```python
# sort
numbers = [4, 1, 3, 2]
numbers.sort()

print(numbers)
```

```python
# reverse
numbers = [1, 2, 3, 4]
numbers.reverse()

print(numbers)
```

## Lists vs dictionaries

Use LISTS when:
* When position matters and you access by position (nodal coordinates)
* You need sequential processing (deflections along the beam)
* Order matters (load steps)

Use DICTIONARIES when:
* You access by name (`materials['S355']`, `sections['UB305']`)
* Clarity is crucial
* You have a lookup table/database

```python
# When position matters - nodal coordinates
node_coords = [(0,0), (5,0), (5,3)]
node_1_position = node_coords[1]  # (5,0)
```

```python
# Sequential results - deflections along a beam
deflections = [0.0, 2.3, 4.1, 3.8, 0.0]  # mm
point_3_deflection = deflections[2]  # 4.1 mm
```

```python
# Ordered collections - load steps in an analysis
load_steps = [0, 10, 20, 30, 40]  # kN
step_2_load = load_steps[2]  # 20 kN
```

```python
# List
beam_list = ["UB305x165x54", 54.0, 30.65, 11700]
print("List approach:")
print(f"Beam mass per metre: {beam_list[1]} kg/m")
print(f"Beam Ix: {beam_list[3]} cm^4\n")

# Dictionary
beam_dict = {
    "name": "UB305x165x54",
    "mass_kg_m": 54.0,
    "depth_cm": 30.65,
    "Ix_cm4": 11700
}
print("Dictionary approach:")
print(f"Beam mass per metre: {beam_dict["mass_kg_m"]} kg/m")
print(f"Beam Ix: {beam_dict["Ix_cm4"]} cm^4\n")
```
