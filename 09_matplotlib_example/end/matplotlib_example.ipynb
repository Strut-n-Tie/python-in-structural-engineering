# Matplotlib example

```python
from matplotlib import pyplot as plt
```

## UDL beam diagrams

<img src="beam_UDL.svg" width="300">

For a simply supported beam under UDL: 
* $V(x) = \dfrac{w L}{2} - w x$
* $M(x) = \dfrac{w L}{2}x - \dfrac{w}{2}x^2 \qquad \text{(sagging positive)}$
* $\theta(x) = \dfrac{w}{24EI}\big(4x^3 - 6Lx^2 + L^3\big)$ 
* $y(x) = \dfrac{w}{24EI}(x^4 - 2Lx^3 + L^3x) \qquad \text{(downwards positive)}$

Maximum values:
* $V(0) = \dfrac{w L}{2}$
* $M(L/2) = \dfrac{w L^2}{8}$
* $\theta(0) = \dfrac{wL^3}{24EI}$
* $y(L/2) = \dfrac{5wL^4}{384EI}$

```python
w = 30E3 # N/m
E = 200E9 # Pa
I = 41000 * 1E-8 # m^4 UB 457 x 191x 89
L = 6
```

```python
def V_x(x): 
    return L*w/2 - w*x
def M_x(x):
    return  L*w*x/2 - w*x**2/2
def theta_x(x): 
    return w*(L**3 - 6*L*x**2 + 4*x**3)/(24*E*I)
def y_x(x): 
    return w*x*(L**3 - 2*L*x**2 + x**3)/(24*E*I)

def V_max(): 
    return L*w/2
def M_max():
    return  L**2*w/8
def theta_max(): 
    return w*L**3/(24*E*I)
def y_max(): 
    return 5*w*L**4/(384*E*I)
```

```python
x_arr = [i/10 for i in range(L*10+1)]
V_arr = [V_x(x)/1000 for x in x_arr]
M_arr = [M_x(x)/1E6 for x in x_arr]
theta_arr = [theta_x(x) for x in x_arr]
y_arr = [y_x(x)*1E3 for x in x_arr]

fig, ax = plt.subplots(4,1,figsize=(7,13))

ax[0].grid(True, alpha=0.3)
ax[0].plot([0, L], [0, 0], color='k', linewidth = 3)
ax[0].plot(x_arr, V_arr, color='C0')
ax[0].plot(0, V_max()/1000, "ro")
ax[0].text(0, V_max()/1000*0.8, "$V_{max}$")
ax[0].set_xlabel("Length [m]")
ax[0].set_ylabel("Shear [kN]")
ax[0].set_xlim([0, L])
ax[0].set_title("Beam shear forces")

ax[1].grid(True, alpha=0.3)
ax[1].plot([0, L], [0, 0], color='k', linewidth = 3)
ax[1].plot(x_arr, M_arr, color='C1')
ax[1].plot(0.5*L, M_max()/1E6, "ro")
ax[1].text(0.5*L, M_max()/1E6*0.9, "$M_{max}$")
ax[1].set_xlabel("Length [m]")
ax[1].set_ylabel("Moment [kNm]")
ax[1].set_xlim([0, L])
ax[1].invert_yaxis() # Downwards positive
ax[1].set_title("Beam bending moments")

ax[2].grid(True, alpha=0.3)
ax[2].plot([0, L], [0, 0], color='k', linewidth = 3)
ax[2].plot(x_arr, theta_arr, color='C2')
ax[2].plot(0, theta_max(), "ro")
ax[2].text(0, theta_max()*0.85, r"$\theta_{max}$")
ax[2].set_xlabel("Length [m]")
ax[2].set_ylabel("Rotation [rad]")
ax[2].set_xlim([0, L])
ax[2].set_title("Beam rotations")

ax[3].grid(True, alpha=0.3)
ax[3].plot([0, L], [0, 0], color='k', linewidth = 3)
ax[3].plot(x_arr, y_arr, color='C3')
ax[3].plot(0.5*L, y_max()*1000, "ro")
ax[3].text(0.5*L, y_max()*1000*0.9, "$y_{max}$")
ax[3].set_xlabel("Length [m]")
ax[3].set_ylabel("Deflection [mm]")
ax[3].set_xlim([0, L])
ax[3].invert_yaxis() # Downwards positive
ax[3].set_title("Beam deflections")

plt.tight_layout()

plt.show()
```
