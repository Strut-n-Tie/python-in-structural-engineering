## Printing

The empty string before the colon means "take the next provided argument to format()" – in this case the x as the only argument.
The 10.4f part after the colon is the format specification.
The f denotes fixed-point notation.
The 10 is the total width of the field being printed, lefted-padded by spaces.
The 4 is the number of digits after the decimal point.

````python
beam_length, E, I = 8, 210E9, 0.025
print(beam_length)
print(E)
print(I)
````

## Operators

| Operators | Syntax  | Functioning       |
|-----------|---------|-------------------|
| `+`       | `x + y` | Addition          |
| `-`       | `x - y` | Subtraction       |
| `*`       | `x * y` | Multiplication    |
| `/`       | `x / y` | Division          |
| `//`      | `x // y`| Quotient          |

````python
# EXPRESSIONS
a, b, c = 1, 2, 3
average=(a+b+c)/3
print(average)

b = 3 ** 2
print(b)

quotient = 5 // 2
remainder = 5 % 2
print(f"quotient = {quotient}, remainder = {remainder}")
````

````python
# Functions

def circle_area(r):
    A = 3.14 * r**2
    return A

radius = 3
print(f"Circle radius = {radius} m; area = {circle_area(radius)} m^2")

def ss_beam_PL_deflection_formula():
    print("Simply supported beam with point load deflection = PL^3/48EI")

ss_beam_PL_deflection_formula()

def ss_beam_PL_deflection():
    return P*L**3/(48*E*I)

# UB 457 x 191 x 67
E = 210E9 # Pa
L = 10 # m
I = 0.0294 # m ^4
P = 300E3 # N

defl = ss_beam_PL_deflection() # m

print(f"Beam deflection = {defl * 1000} mm")
````

## Logical operators

| Operator | Syntax                     | Result                                                                 |

| `and`    | `A and B` | • `True` if both `A` and `B` are `True`<br>• `False` otherwise |

## IF STATEMENTS

Example use:
```
if A:
    do something
elif B and C: (optional)
    do something else
elif D or E: (optional)
    do something different
else: (optional)
    do this
```

````python
deflection = 23 # mm

if deflection <= 15:
    print("Immediate Occupancy") 
elif deflection > 15 and deflection <= 40:
    print("Life Safety") 
else:
    print("Collapse prevention") 
````
