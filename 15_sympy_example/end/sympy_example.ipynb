# SymPy

```python
import sympy as sp
```

```python
# Define symbols
x, L = sp.symbols('x L', positive=True)
E, I, P = sp.symbols('E I P', positive=True)
w0 = sp.symbols('w0')
```

```python
# Buckling Mode Shape (Pinned–Pinned)
w = w0 * sp.sin(sp.pi * x / L)
w
```

```python
# This choice automatically satisfies boundary conditions:
# w(0)=0
# w(L)=0

# You can confirm symbolically:
sp.simplify(w.subs(x, 0)), sp.simplify(w.subs(x, L))
```

```python
# Rotation (Slope) Using Symbolic Differentiation
theta = sp.diff(w, x)
sp.simplify(theta)
```

```python
# Curvature (Second Derivative)
curvature = sp.diff(w, x, 2)
sp.simplify(curvature)
```

```python
# Internal Bending Moment
M = E * I * curvature
sp.simplify(M)
```

```python
# Governing buckling equation
buckling_eq = sp.Eq(M, -P * w)
buckling_eq
```

```python
# Solve
Pcr = sp.solve(
    sp.factor(E * I * (sp.pi / L)**2 - P),
    P
)[0]
```
