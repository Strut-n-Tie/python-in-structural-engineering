## Section resistance

$F_{Rd} = \sigma * A$

$M_{Rd,y} = W_{yy} * \sigma$

```python
sigma = 355E6 # Pa
```

```python
UB_sections
```

```python
UB_resistance = UB_sections[["Section"]].copy()
UB_resistance["F_Rd"] = UB_sections["A"] * 1e-4 * sigma
UB_resistance["M_yy_Rd"] = UB_sections["Wyy"] * 1e-6 * sigma
UB_resistance["M_zz_Rd"] = UB_sections["Wzz"] * 1e-6 * sigma
UB_resistance["Section"] = "UB" + UB_sections["Section"].astype(str)
UB_resistance
```

```python
UC_resistance = UC_sections[["Section"]].copy()
UC_resistance["F_Rd"] = UC_sections["A"] * 1e-4 * sigma
UC_resistance["M_yy_Rd"] = UC_sections["Wyy"] * 1e-6 * sigma
UC_resistance["M_zz_Rd"] = UC_sections["Wzz"] * 1e-6 * sigma
UC_resistance["Section"] = "UC" + UC_resistance["Section"].astype(str)
```

```python
resistances = pd.concat([UB_resistance,UC_resistance]).reset_index(drop=True)
resistances
```

```python
# Add line column
combined_df = load_combination.copy()

combined_df = combined_df.merge(
    elements[["Element", "Line"]],
    on="Element",
    how="left"
)
combined_df
```

```python
# Add section column
combined_df = combined_df.merge(
    line_section[["line", "section"]],
    left_on="Line",
    right_on="line",
    how="left"
)
combined_df = combined_df.drop(columns=["line"])
combined_df
```

```python
# Add resistances
combined_df = combined_df.merge(
    resistances[["Section", "F_Rd","M_yy_Rd","M_zz_Rd"]],
    left_on="section",
    right_on="Section",
    how="left"
)
combined_df = combined_df.drop(columns=["Section"])
combined_df
```

```python
# Add utilisation column
combined_df["utilisation"] = (
    (abs(combined_df["Fx[N]"]) / combined_df["F_Rd"]) +
    (abs(combined_df["My[N.m]"]) / combined_df["M_yy_Rd"]) +
    1.5 * (abs(combined_df["Mz[N.m]"]) / combined_df["M_zz_Rd"])
)
combined_df
```

```python
# Get maximum utilisation
print(f"max utilisation = {max(combined_df["utilisation"])}")
combined_df.loc[ combined_df["utilisation"] == max(combined_df["utilisation"]) ]
```
