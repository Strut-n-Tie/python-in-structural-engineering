# Numpy

```python
import numpy as np
```

```python

# 3DOF shear building (lumped masses, story stiffnesses)
masses = np.array([200e3, 200e3, 200e3])
k_story = np.array([150e6, 150e6, 150e6])

M = np.diag(masses)
K = np.array([[ k_story[0]+k_story[1], -k_story[1],             0         ],
              [ -k_story[1],           k_story[1]+k_story[2],  -k_story[2]],
              [ 0,                    -k_story[2],             k_story[2] ]], dtype=float)

# Solve generalized eigenproblem: K φ = ω² M φ
w2, Phi = np.linalg.eigh(np.linalg.inv(M) @ K)
wn = np.sqrt(w2)                   # natural frequencies
# Mass-normalize modes
for i in range(3):
    mi = Phi[:, i].T @ M @ Phi[:, i]
    Phi[:, i] /= np.sqrt(mi)

# Participation factors for base acceleration (Γ = φ^T M 1)
ones = np.ones(3)
Gamma = Phi.T @ M @ ones

print("Natural periods [s]:", 2*np.pi / wn)
print("Participation factors:", Gamma)
```
